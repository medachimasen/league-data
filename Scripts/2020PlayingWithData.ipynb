```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Raw/2020_LoL_esports_match_data_from_OraclesElixir_20200821.csv')
dfLCK = df[df.league == 'LCK']
dfLPL = df[df.league == 'LPL']
dfLCS = df[df.league == 'LCS']
dfLEC = df[df.league == 'LEC']
dfLCKSpring = dfLCK[dfLCK.split == 'Spring']
dfLCKSummer = dfLCK[dfLCK.split == 'Summer']
dfLPLSpring = dfLPL[dfLPL.split == 'Spring']
dfLPLSummer = dfLPL[dfLPL.split == 'Summer']
dfLCSSpring = dfLCS[dfLCS.split == 'Spring']
dfLCSSummer = dfLCS[dfLCS.split == 'Summer']
dfLECSpring = dfLEC[dfLEC.split == 'Spring']
dfLECSummer = dfLEC[dfLEC.split == 'Summer']
```

```python
dfLPLJDGSummergames = dfLPLSummer[dfLPLSummer.team == 'JD Gaming']
test=dfLPLJDGSummergames.gameid.unique()
dfconcat = pd.DataFrame()
for x in test:
    dfconcat = dfconcat.append(dfLPLSummer[dfLPLSummer.gameid == x])
###This gets all games with JDG
```
